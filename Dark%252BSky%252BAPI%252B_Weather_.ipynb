## 5) Go through the daily forecast, printing out the next week's worth of predictions.

I'd like to know the high temperature for each day, and whether it's hot, warm, or cold (based on what temperatures you think are hot, warm or cold).

```python
daily = data['daily']
for day_to_day in daily['data']:   
    if day_to_day['temperatureMax']>=75:
        print("hot")
        print(day_to_day['temperatureMax'])
    if day_to_day['temperatureMax']<75:
        print("cold")
        print(day_to_day['temperatureMax'])
```

## 6) What's the weather looking like for the rest of today in Miami, Florida?

I'd like to know the temperature for every hour, and if it's going to have cloud cover of more than 0.5 say "{temperature} and cloudy" instead of just the temperature.

```python
result = requests.get('https://api.darksky.net/forecast/c9e5790066c1ca03c73295f03f577372/25.7617,80.1918')
```

```python
data2 = result.json()
```

```python
print(data2.keys())
print(data2['hourly'])
```

```python
import datetime
import time
datetime.datetime.fromtimestamp(int("1487728800"))

time.strftime("%D %H:%M", time.localtime(int("1284101485")))
```

```python
for hour in data2['hourly']['data']:
    if hour["cloudCover"]>0.5:
        print(time.strftime('%D %H:%M', time.localtime(int(hour['time']))), ", temp: ", str(hour['temperature']), ", also cloudy")
    else:
        print(time.strftime('%D %H:%M', time.localtime(int(hour['time']))), ", temp: ", str(hour['temperature']))
```

## 7) What was the temperature in Central Park on Christmas Day, 1980? How about 1990? 2000?
